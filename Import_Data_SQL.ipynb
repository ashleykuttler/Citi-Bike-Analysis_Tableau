```python
# CREATE CONNECTION STRINGS TO DATABASE
rds_connection_string = f'root:{password}@127.0.0.1/citi_bike_db'
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
#DROP EXISTING TABLE BEFORE ADDING NEW DATA
with engine.connect() as con:
    con.execute("DROP TABLE citibike_tripdata")


#WRITE TRIP DATA CSVS TO PANDAS DATAFRAMES, IMPORT INTO MYSQL
#MARCH 2019
tripdata_201903 = pd.DataFrame(pd.read_csv('https://s3.amazonaws.com/tripdata/JC-201903-citibike-tripdata.csv.zip'))
tripdata_201903.to_sql(name='citibike_tripdata', con=engine, if_exists='append', index=False)
#FEB 2019
tripdata_201902 = pd.DataFrame(pd.read_csv('https://s3.amazonaws.com/tripdata/JC-201902-citibike-tripdata.csv.zip'))
tripdata_201902.to_sql(name='citibike_tripdata', con=engine, if_exists='append', index=False)
#JAN 2019
tripdata_201901 = pd.DataFrame(pd.read_csv('https://s3.amazonaws.com/tripdata/JC-201901-citibike-tripdata.csv.zip'))
tripdata_201901.to_sql(name='citibike_tripdata', con=engine, if_exists='append', index=False)


#MARCH 2018
tripdata_201803 = pd.DataFrame(pd.read_csv('https://s3.amazonaws.com/tripdata/JC-201803-citibike-tripdata.csv.zip'))
tripdata_201803.to_sql(name='citibike_tripdata', con=engine, if_exists='append', index=False)
#FEB 2018
tripdata_201802 = pd.DataFrame(pd.read_csv('https://s3.amazonaws.com/tripdata/JC-201802-citibike-tripdata.csv.zip'))
tripdata_201802.to_sql(name='citibike_tripdata', con=engine, if_exists='append', index=False)
#JAN 2018
tripdata_201801 = pd.DataFrame(pd.read_csv('https://s3.amazonaws.com/tripdata/JC-201801-citibike-tripdata.csv.zip'))
tripdata_201801.to_sql(name='citibike_tripdata', con=engine, if_exists='append', index=False)


```
